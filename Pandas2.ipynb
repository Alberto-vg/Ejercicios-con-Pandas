```python
# graficos_plot
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV previamente creado
df = pd.read_csv("datos_personas.csv")

# Mostrar los primeros datos
print(df.head())


# 1. Gráfico de barras (bar plot) del sexo

conteo_sexo = df["SEXO"].value_counts()

plt.bar(conteo_sexo.index, conteo_sexo.values, color=["pink", "lightblue"])
plt.title("Número de personas por sexo")
plt.xlabel("Sexo")
plt.ylabel("Frecuencia")
plt.show()


# 2. Boxplot del peso con respecto al sexo

# Separar pesos por sexo
pesos_mujeres = df[df["SEXO"] == "F"]["PESO"]
pesos_hombres = df[df["SEXO"] == "M"]["PESO"]

plt.boxplot([pesos_mujeres, pesos_hombres],
            tick_labels=["Mujeres", "Hombres"],
            patch_artist=True)

plt.title("Boxplot de peso por sexo")
plt.ylabel("Peso (kg)")
plt.show()
```
